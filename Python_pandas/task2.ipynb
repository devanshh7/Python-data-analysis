`````python
# Create a MESSY version of the dataset for cleaning practice
import pandas as pd
messy_data = {
    'order_id': ['O001','O002','O002','O003',None,'O005'],       # duplicate + null
    'customer_name': ['  Ravi Kumar ','PRIYA SHARMA','priya sharma','Amit Singh','Sneha',''],
    'amount': [15000, None, 25000, -500, 32000, 'N/A'],           # null, negative, string
    'email': ['ravi@gmail.com','PRIYA@GMAIL.COM','priya@gmail.com','amit@','sneha@gmail.com',''],
    'phone': ['9876543210', '987-654-3210', '98765', '9876543212', None, '9876543214'],
    'city': ['mumbai','DELHI','Delhi','Bangalore ','  Chennai','pune'],
}
df = pd.DataFrame(messy_data)
print (df)
`````

`````python
# Q1. Find all null values — show count per column

n_count= df.isnull().sum()
print(f'Null values :\n{n_count}')
`````

`````python

# See duplicate rows (based on order_id)
duplicate_rows = df[df.duplicated(subset=['order_id'], keep=False)]
print(f'Duplicate rows:\n{duplicate_rows}')

# Q2. Drop duplicate rows (based on order_id)
df_no_duplicates = df.drop_duplicates(subset=['order_id'])
print(f'DataFrame after dropping duplicates:\n{df_no_duplicates}')
`````

`````python
# Q3. Clean customer_name: strip spaces + Title Case
df['customer_name']= df['customer_name'].str.strip().str.title()
print(f'Cleaned customer names:\n{df["customer_name"]}')
`````

`````python
# Q4. Standardize city: strip + Title Case
df['city']= df['city'].str.strip().str.title()
print(f'Cleaned City names:\n{df["city"]}')
`````

`````python
result= df[pd.to_numeric(df['amount'], errors='coerce').ge(0)]
print(f'Rows with invalid amount:\n{result}')
`````

`````python
# Q6. Standardize email to lowercase

df['email']= df['email'].str.lower()
print(f'Lowercase emails:\n{df["email"]}')
`````

`````python
df['invalid_mail']= df['email'].str.contains('@') & df['email'].str.contains('.')
print(df[['invalid_mail','email']])
`````

`````python
# Q8. Fill null phone with "Not Provided"
df['phone']= df['phone'].fillna('Not Provided')
print(f'Phone column after filling nulls:\n{df["phone"]}')
`````

`````python
# Q9. Reset the index after cleaning
df.reset_index(drop=True, inplace=True)
print(f'DataFrame after resetting index:\n{df}')
`````

`````python
print(df.dropna(subset=['order_id'], inplace=True))
#replace the empty string with pd.NA to mark it as null, then drop rows where customer_name is null
df.replace({'customer_name': {'': pd.NA}}, inplace=True)
print(df.dropna(subset=['customer_name'], inplace=True))
print(f'DataFrame after dropping rows with null order_id:\n{df}')
`````
